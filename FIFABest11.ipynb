```python
#weights #BEST_GOALKEEPER
a = 0.5
b = 1
c = 2
d = 3
```

```python
#BEST_GOALKEEPER Characteristics

df['gk_Shot_Stopper']=(a*df.Reactions + b*df.Composure + a*df.Speed + a*df.Strength + c*df.Jumping + b*df.GK_Positioning + c*df.GK_Diving + d*df.GK_Reflexes + b*df.GK_Handling)/(2*a + 4*b + 2*c + 1*d)
df['gk_Sweeper']=(b*df.Reactions + b*df.Composure + b*df.Speed + a*df.Short_Pass + a*df.Long_Pass + b*df.Jumping + b*df.GK_Positioning + b*df.GK_Diving + b*df.GK_Reflexes + b*df.GK_Handling + d*df.GK_Kicking + c*df.Vision)/(2*a + 9*b + 1*c + 1*d)
```

```python
df.columns
```

```python
plt.figure(figsize=(20,10))
sd1=df.sort_values('gk_Shot_Stopper',ascending=False)[:5]
x1=np.array(list(sd1['Name']))
y1=np.array(list(sd1['gk_Shot_Stopper']))

sns.barplot(x1,y1,palette='colorblind')
plt.ylabel('shot stopping score')



```

```python
plt.figure(figsize=(20,10))
sd1=df.sort_values('gk_Sweeper',ascending=False)[:5]
x1=np.array(list(sd1['Name']))
y1=np.array(list(sd1['gk_Sweeper']))

sns.barplot(x1,y1,palette='colorblind')
plt.ylabel('sweeping score')
```

```python
#Best Defender

df['centre_backs']=(d*df.Reactions + c*df.Interceptions + d*df.Sliding_Tackle + d*df.Sliding_Tackle + d*df.Standing_Tackle + b*df.Vision + c*df.Speed + d*df.Long_Pass + c*df.Short_Pass + b*df.Heading + a*df.Volleys )/(1*a+2*b + 3*c + 5*d)

df['wing_backs']=(d*df.Reactions + c*df.Interceptions + d*df.Sliding_Tackle + d*df.Sliding_Tackle + d*df.Standing_Tackle + a*df.Vision + c*df.Speed + d*df.Long_Pass + d*df.Short_Pass + c*df.Shot_Power)/( 1*b + 3*c + 5*d)
```

```python

#left center back
plt.figure(figsize=(15,6))
sd=df[(df['Club_Position']=='LCB')].sort_values('centre_backs',ascending=False)[:5]
x2=np.array(list(sd['Name']))
y2=np.array(list(sd['centre_backs']))
sns.barplot(x2,y2, palette=sns.color_palette("plasma"))
plt.ylabel("LCB Score")
```

```python
#MID FIELDER

df['controller']=(b*df.Weak_foot + b*df.Skill_Moves + a*df.Dribbling + d*df.Ball_Control + a*df.Marking + c*df.Vision + d*df.Composure + d*df.Short_Pass + d*df.Long_Pass)/ (2*a+1*c+2*b+4*d)

df['playmaker']=(d*df.Weak_foot + c*df.Skill_Moves + d*df.Dribbling + d*df.Ball_Control + b*df.Marking + a*df.Vision + d*df.Composure + b*df.Short_Pass + d*df.Long_Pass + d*df.Attacking_Position + c*df.Crossing)/ (1*a+2*b+2*c+6*d)

df['mf_beast']=(b*df.Weak_foot + d*df.Skill_Moves + c*df.Dribbling + c*df.Ball_Control + b*df.Marking + b*df.Vision + d*df.Composure + c*df.Short_Pass + c*df.Long_Pass)/ (1*a+3*b+4*c+3*d)
```

```python
#plot controller midfielder

plt.figure(figsize=(15,6))
ss =df[(df['Club_Position'] == 'LCM')].sort_values('controller', ascending=False)[:5]
x1=np.array(list(ss['Name']))
y1=np.array(list(ss['controller']))
sns.barplot(x1,y1,palette=sns.diverging_palette(145,280,s=85,l=25,n=5))
plt.ylabel("Controller score")
```

```python
#plot playmaker midfielder

plt.figure(figsize=(15,6))
ss =df[(df['Club_Position'] == 'CAM') | (df['Club_Position'] =="LAM") | (df['Club_Position']== 'RAM')].sort_values('playmaker', ascending=False)[:5]
x1=np.array(list(ss['Name']))
y1=np.array(list(ss['playmaker']))
sns.barplot(x1,y1,palette=sns.diverging_palette(145,280,s=85,l=25,n=5))
plt.ylabel("playmaker")
```

```python
#plot beast midfielder


plt.figure(figsize=(15,6))
ss =df[(df['Club_Position'] == 'RCM') | (df['Club_Position'] =="RAM")].sort_values('mf_beast', ascending=False)[:5]
x2=np.array(list(ss['Name']))
y2=np.array(list(ss['mf_beast']))
sns.barplot(x2,y2,palette=sns.diverging_palette(145,280,s=85,l=25,n=5))
plt.ylabel("beast score")
```

```python
#BEST ATTACKER

df['left_attack']=(c*df.Weak_foot + c*df.Ball_Control + d*df.Dribbling + d*df.Aggression + c*df.Attacking_Position + c*df.Speed + b*df.Short_Pass + a*df.Long_Pass + d*df.Finishing + b*df.Penalties + d*df.Freekick_Accuracy + c*df.Heading + d*df.Skill_Moves)/(c*5+d*5+1*a+2*b)
df['right_attack']=(c*df.Weak_foot + c*df.Ball_Control + d*df.Dribbling + d*df.Aggression + c*df.Attacking_Position + c*df.Speed + b*df.Short_Pass + a*df.Long_Pass + d*df.Finishing + d*df.Penalties + d*df.Freekick_Accuracy + c*df.Heading + d*df.Skill_Moves)/(c*5+d*5+1*a+2*b)
df['striker_attack']=(b*df.Weak_foot + a*df.Balance+ d*df.Ball_Control + d*df.Dribbling + d*df.Aggression + c*df.Attacking_Position + c*df.Speed + b*df.Short_Pass + a*df.Long_Pass + d*df.Finishing + d*df.Penalties + d*df.Freekick_Accuracy + c*df.Heading)/(c*3+b*2+d*6+a*2) 

```

```python
plt.figure(figsize=(15,6))
ss =df[(df['Club_Position'] == 'LW') | (df['Club_Position'] =="LM") | (df['Club_Position']== 'LS')].sort_values('left_attack', ascending=False)[:5]
x1=np.array(list(ss['Name']))
y1=np.array(list(ss['left_attack']))
sns.barplot(x1,y1,palette=sns.diverging_palette(255,133,l=60,n=5, center='dark'))
plt.ylabel("LEFT ATTACK PLAYER")
```
